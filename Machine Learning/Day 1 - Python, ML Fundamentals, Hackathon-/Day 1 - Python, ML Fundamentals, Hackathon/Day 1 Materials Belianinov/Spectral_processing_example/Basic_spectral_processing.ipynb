### Spectral Processing Example

```python
from sklearn import preprocessing #data processing module
import numpy as np #fast math
from matplotlib import pyplot as plt #plotting utilities 
%matplotlib inline 
plt.rcParams['figure.figsize'] = [20, 10] #figure size

sample = np.array([0, 1.0, 0, 2, 2, 2, 3, 5, 7, 7 , 8, 10, 12, 11, 9, 8, 8, 7, 5, 3, 3, 1, 9, 9, 3, 1, 0, 0]) #array

plt.plot(sample) 
```

```python
sample_scaled = preprocessing.scale(sample) #zero mean and unit variance

plt.plot(sample, 'r') 
plt.plot(sample_scaled, 'g')
```

### Scale a toy dataset to the [0, 1] range

```python
sample_array = np.tile(sample, [5, 1])

print(np.shape(sample_array))

sample_array = np.transpose(sample_array, [1, 0])

print(np.shape(sample_array))

min_max_scaler = preprocessing.MinMaxScaler()
sample_train_minmax = min_max_scaler.fit_transform(sample_array)

plt.plot(sample, 'r') 
plt.plot(sample_train_minmax[:, 1], 'g') #processed data
```
